# Govt dataset analysis
# CROP Total Food grains (Area in Hectares, Production in Tones, Yield in Kgs/Hectare)
# Dataset datafile.csv


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path = "datafile.csv" 

df = pd.read_csv(file_path)
df

```

 # Display basic information and first few rows

```python
 display(df.info())
display(df.head())
```

# Handling missing values

```python
 display(df.isnull().sum())  
df.dropna(inplace=True)   
```

# Data Cleaning

```python
 
df.drop_duplicates(inplace=True)


df.columns = df.columns.str.strip().str.replace(' ', '_').str.lower()
df
```

# Converting numerical columns to appropriate data types 

```python
num_cols = df.select_dtypes(include=['object']).columns
df[num_cols] = df[num_cols].apply(pd.to_numeric, errors='coerce')
```


# Fill missing values with mean (if applicable)


```python
df.fillna(df.mean(), inplace=True)
```

# Summary statistics

```python
display(df.describe())
```


# Visualization: Distribution of Kharif Yield

```python
plt.figure(figsize=(10,5))
sns.histplot(df['kharif_yield'], bins=20, kde=True)
plt.title('Distribution of Kharif Yield')
plt.xlabel('Kharif Yield')
plt.ylabel('Frequency')
plt.show()
```


# Top 5 districts by All Seasons Production

```python

top_production_districts = df[['district_name', 'all_seasons_production']].sort_values(by='all_seasons_production', ascending=False).head()
display(top_production_districts)
```


# Correlation heatmap

```python

plt.figure(figsize=(12,6))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```

#   Visualizations


```python
 plt.figure(figsize=(10,5))
sns.boxplot(x=df['kharif_yield'])
plt.title('Box Plot of Kharif Yield')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(x=df['rabi_yield'])
plt.title('Box Plot of Rabi Yield')
plt.show()
```

```python

plt.figure(figsize=(12,6))
sns.barplot(x='district_name', y='all_seasons_production', data=df.sort_values(by='all_seasons_production', ascending=False).head(10))
plt.xticks(rotation=45)
plt.title('Top 10 Districts by All Seasons Production')
plt.show()
```

```python

plt.figure(figsize=(12,6))
sns.scatterplot(x='kharif_areaafter_bund_correction_factor', y='kharif_production', data=df)
plt.title('Kharif Area vs Production')
plt.xlabel('Kharif Area After Bund Correction')
plt.ylabel('Kharif Production')
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.scatterplot(x='rabi_areaafter_bund_correction_factor', y='rabi_production', data=df)
plt.title('Rabi Area vs Production')
plt.xlabel('Rabi Area After Bund Correction')
plt.ylabel('Rabi Production')
plt.show()
```

```python

plt.figure(figsize=(12,6))
sns.lineplot(x='district_name', y='all_seasons_yield', data=df.sort_values(by='all_seasons_yield', ascending=False))
plt.xticks(rotation=45)
plt.title('All Seasons Yield by District')
plt.show()
```

```python


plt.figure(figsize=(10,5))
sns.violinplot(x=df['summer_yield'])
plt.title('Violin Plot of Summer Yield')
plt.show()
```

```python

plt.figure(figsize=(12,6))
sns.pairplot(df[['kharif_yield', 'rabi_yield', 'summer_yield', 'all_seasons_yield']])
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.countplot(y='district_name', data=df, order=df['district_name'].value_counts().index[:10])
plt.title('Top 10 Districts by Frequency in Dataset')
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.kdeplot(df['all_seasons_production'], shade=True)
plt.title('Kernel Density Estimate of All Seasons Production')
plt.show()
```

```python

plt.figure(figsize=(12,6))
sns.regplot(x='all_seasons_yield', y='all_seasons_production', data=df)
plt.title('Regression Plot: All Seasons Yield vs Production')
plt.show()
```
